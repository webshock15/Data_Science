```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

# Chaper1. Classic Graphs

## Classic Graph1 : barbell_graph

```python
classic_graph1 = nx.barbell_graph(10,10)
nx.draw(classic_graph1,node_size=500,with_labels=True,font_size=12)
```

```python
radius1 = nx.radius(classic_graph1)
diameter1 = nx.diameter(classic_graph1)
C_node1 = nx.center(classic_graph1)
P_node1 = nx.periphery(classic_graph1)
coef_of_C1 = nx.clustering(classic_graph1)
coef_of_AC1 = nx.average_clustering(classic_graph1)
```

```python
print "Classic Graph1 : barbell_graph"
print "radius:", radius1 
print "diameter:", diameter1
print "Center node:", C_node1
print "Periphery node:", P_node1
print "Each nodes of Clustering Coef:",coef_of_C1
print "Average Clustering Coef:",coef_of_AC1
```

#### Remark
- 각 노드들의 Clustering coefficient를 살펴보면 양쪽으로 0~8노드, 21~29번노드가 각각 9개씩, 그리고 양쪽 노드의 시작으로 9번노드와 22번 노드가 있음을 알 수 있음.
- Average clustering coefficient를 통해서 어느정도 집단화 되어 있음을 알 수 있으며, 위의 Clustering Coef를 통해서 센터 노드인 14,15 노드의 양쪽 끝에서 집단화 되었음을 유추할 수 있음.
- 위의 계수들과 center node, periphery node를 통해서 14,15번 노드를 중심으로 0~8노드, 21~29번노드가 뭉쳐 있음을 유추할 수 있다.
